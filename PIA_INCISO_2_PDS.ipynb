```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Parámetros de la señal simulada de EEG
fs = 1000  # Frecuencia de muestreo en Hz
t_simulacion = 10  # Duración de la simulación en segundos
t = np.arange(0, t_simulacion, 1/fs)  # Vector de tiempo

# Componentes de frecuencia (ondas alfa y beta)
alfa = 10  # Frecuencia en Hz
beta = 20  # Frecuencia en Hz

# Señal simulada de EEG (combinación de ondas alfa y beta con ruido)
eeg_signal = 0.5 * np.sin(2 * np.pi * alfa * t) + 0.3 * np.sin(2 * np.pi * beta * t) + np.random.normal(0, 0.2, len(t))

# Aplica el método de Welch a la señal de EEG
frequencies, power_spectrum = welch(eeg_signal, fs=fs, nperseg=256, noverlap=128)

# Visualización de la señal simulada de EEG y su espectro de potencia
plt.figure(figsize=(12, 6))

# Señal de EEG
plt.subplot(2, 1, 1)
plt.plot(t, eeg_signal)
plt.title('Señal Simulada de EEG')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

# Espectro de potencia utilizando el método de Welch
plt.subplot(2, 1, 2)
plt.semilogy(frequencies, power_spectrum)
plt.title('Estimación del Espectro de Potencia (Método de Welch)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Densidad de Potencia Espectral')
plt.grid(True)

plt.tight_layout()
plt.show()
```
